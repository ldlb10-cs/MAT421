20.1 Numerical Differentiation Problem Statement

*   Numerical differentiation is a technique used to approximate the derivative of a function when an analytical derivative is difficult or impossible to obtain. It is commonly used in engineering, physics, and data analysis to estimate the rate of change of a function based on discrete data points.



```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)

def forward_difference(f, x, h):
    return (f(x + h) - f(x)) / h

def backward_difference(f, x, h):
    return (f(x) - f(x - h)) / h

def central_difference(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

x_vals = np.linspace(0, 2 * np.pi, 100)
h = 0.1

fwd_diff = forward_difference(f, x_vals, h)
bwd_diff = backward_difference(f, x_vals, h)
ctr_diff = central_difference(f, x_vals, h)

true_derivative = np.cos(x_vals)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(x_vals, true_derivative, label="True Derivative (cos(x))", linestyle="dashed")
plt.plot(x_vals, fwd_diff, label="Forward Difference", alpha=0.7)
plt.plot(x_vals, bwd_diff, label="Backward Difference", alpha=0.7)
plt.plot(x_vals, ctr_diff, label="Central Difference", alpha=0.7)
plt.legend()
plt.xlabel("x")
plt.ylabel("Derivative Approximation")
plt.title("Numerical Differentiation using Finite Differences")
plt.grid()
plt.show()
```

20.3 Approximating of Higher Order Derivatives

*   Approximating higher-order derivatives involves estimating the derivatives of a function beyond the first and second order using finite difference methods. These methods provide numerical approximations when analytical differentiation is difficult or when the function is only known at discrete points.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2 * np.pi, 0.01)

omega = 100
epsilon = 0.02

y = np.sin(x)
y_noise = y + epsilon * omega * np.cos(omega * x)

plt.figure(figsize=(8, 6))
plt.plot(x, y_noise, 'r', label="Derivative sin(x) + noise")
plt.plot(x, y, 'b', label="Derivative of sin(x)")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Numerical Differentiation with Noise")
plt.grid()
plt.show()
```
