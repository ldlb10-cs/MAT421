 17.1 Interpolation


*   Interpolation is a method of estimating unknown values that fall between known data points.

*   It is commonly used in data analysis, numerical simulations, and engineering.




```python
import numpy as np
import matplotlib.pyplot as plt

x_data = [0, 1, 2, 3, 4]
y_data = [0, 2, 4, 6, 8]

plt.figure(figsize=(8, 5))
plt.scatter(x_data, y_data, color='red', label='Data Points')
plt.title("Data Points for Interpolation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

17.2 Linear Interpolation


*   Linear Interpolation connects two adjacent points with a straight line and estimates intermediate values.






```python
def linear_interpolation(x0, y0, x1, y1, x):
    return y0 + ((x - x0) / (x1 - x0)) * (y1 - y0)

x0, y0 = 1, 2
x1, y1 = 2, 4
x_interp = 1.5
y_interp = linear_interpolation(x0, y0, x1, y1, x_interp)

plt.figure(figsize=(8, 5))
x_range = np.linspace(x0, x1, 100)
y_range = linear_interpolation(x0, y0, x1, y1, x_range)
plt.plot(x_range, y_range, label='Linear Interpolation', color='blue')
plt.scatter([x0, x1], [y0, y1], color='red', label='Data Points')
plt.scatter(x_interp, y_interp, color='green', label=f'Interpolated Point ({x_interp}, {y_interp:.2f})')
plt.title("Linear Interpolation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

17.3 Cubic Spline Interpolation

*   Cubic Spline Interpolation fits a piecewise cubic polynomial between each pair of data points.




```python
from scipy.interpolate import CubicSpline

x_data = [0, 1, 2, 3, 4]
y_data = [0, 2, 0, 2, 4]
cs = CubicSpline(x_data, y_data)

x_interp = np.linspace(min(x_data), max(x_data), 100)
y_interp = cs(x_interp)

plt.figure(figsize=(8, 5))
plt.plot(x_interp, y_interp, label='Cubic Spline Interpolation', color='blue')
plt.scatter(x_data, y_data, color='red', label='Data Points')
plt.title("Cubic Spline Interpolation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```
