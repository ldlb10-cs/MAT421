1.1 : Introduction

*   Linear Algebra is a fundamental tool in data science and engineering.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eig, qr
from sklearn.linear_model import LinearRegression

vector_a = np.array([2, 4, -3])
vector_b = np.array([-1, 0, 5])
print("Vector A:", vector_a)
print("Vector B:", vector_b)

cross_product = np.cross(vector_a, vector_b)
print("Cross Product:", cross_product)
```

1.2 : Linear Algebra Concepts

*   Linear algebra deals with vector spaces and linear mappings between them.
*   It is fundamental for solving systems of linear equations, transformations, and understanding geometric structures.



```python
A = np.array([[2, 1], [0, 3]])
B = np.array([[1, 4], [2, 5]])
print("Matrix Product:\n", np.dot(A, B))

def eigen_analysis(matrix):
    values, vectors = eig(matrix)
    print("Eigenvalues:", values)
    print("Eigenvectors:\n", vectors)

eigen_analysis(np.array([[3, 1], [0, 2]]))
```

1.3 : Linear Regression

*   Linear regression finds a relationship between a dependent and independent variable



```python
x = np.linspace(0, 10, 50).reshape(-1, 1)
y = 2.5 * x + np.random.randn(50, 1)

model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

plt.scatter(x, y, label='Data')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression Example")
plt.show()

print(f"Slope: {model.coef_[0][0]}")
print(f"Intercept: {model.intercept_[0]}")
```
