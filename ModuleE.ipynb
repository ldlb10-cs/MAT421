3.1 Continuity and Differentiation

*   A function is continuous if its graph has no breaks, jumps, or holes. This means that small changes in the input result in small, predictable changes in the output.
*   A function is differentiable if it has a well-defined slope at every point in its domain. This means that the function changes smoothly without sharp corners or abrupt direction shifts.




```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 400)
y = np.abs(x)

plt.figure(figsize=(6,4))
plt.plot(x, y, label=r"$f(x) = |x|$")
plt.axvline(0, color='r', linestyle='--', label="Not differentiable at x=0")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Continuity and Differentiability of |x|")
plt.legend()
plt.grid()
plt.show()
```



*   Taylor’s theorem provides a way to approximate a complex function using a polynomial expansion around a specific point. This approximation is constructed using the function’s value, slope, and higher-order changes at that point.


```python
x_taylor = np.linspace(-2, 2, 400)
y_exact = np.exp(x_taylor)
y_taylor_2 = 1 + x_taylor + (x_taylor**2)/2
y_taylor_4 = 1 + x_taylor + (x_taylor**2)/2 + (x_taylor**3)/6 + (x_taylor**4)/24

plt.figure(figsize=(6,4))
plt.plot(x_taylor, y_exact, label="Actual $e^x$", linewidth=2)
plt.plot(x_taylor, y_taylor_2, linestyle="--", label="Taylor (2nd order)")
plt.plot(x_taylor, y_taylor_4, linestyle="--", label="Taylor (4th order)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Taylor Series Approximation of $e^x$")
plt.legend()
plt.grid()
plt.show()
```

3.2 Optimization with Gradient Descent

*   Gradient descent is an iterative method used to find the minimum value of a function. It works by calculating the slope (gradient) at a given point and updating the function’s input in the direction that decreases the function’s value. This process is repeated until a point is reached where further updates produce minimal change, indicating an optimal or near-optimal solution.



```python
def gradient_descent(starting_x, learning_rate, iterations):
    x_values = [starting_x]
    x = starting_x
    for _ in range(iterations):
        grad = 2 * x
        x = x - learning_rate * grad
        x_values.append(x)
    return x_values

x_vals = gradient_descent(starting_x=5, learning_rate=0.1, iterations=10)
iterations = np.arange(len(x_vals))

plt.figure(figsize=(6,4))
plt.plot(iterations, x_vals, marker='o', linestyle='-', label="x values during GD")
plt.xlabel("Iterations")
plt.ylabel("x value")
plt.title("Gradient Descent for $f(x) = x^2$")
plt.legend()
plt.grid()
plt.show()
```
